```python
from transformers import AutoTokenizer
import json
import numpy as np
```

```python
data_path = "../data/raw/Subtask_1_train.json"
with open(data_path, "r") as file:
    data = json.load(file)
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
print(tokenizer.get_vocab())
```

```python
d0 = data[0]
```

```python
c0 = d0["conversation"]
```

```python
# tokens = []
# for u in c0:
#     tokens.append(tokenizer(u["text"]))

tokens = tokenizer([u['text'] for u in c0])
tokens
```

```python


tokenized_data_path = "../data/tokenized/train.npy"
tokenized = []
for conversation in data:
    tokens = tokenizer([u['text'] for u in conversation["conversation"]])
    tokenized.append(tokens["input_ids"])
```

```python
np.save(tokenized_data_path, np.array(tokenized))
```
